```python
# assigning a function to a variable

def print_hi():
    print("hi")
    
greet = print_hi

print_hi()
greet()

print(greet)
print(type(greet))
```

```python
# functions nested in other functions

def calculator(a, b):
    def add(a, b):
        return a + b

    def subtract(a, b):
        return a - b

    def multiply(a, b):
        return a * b

    def divide(a, b):
        return a / b
    
    operator = input("Choose an operation - add/subtract/multiply/divide: ")
    if operator == "add":
        return add(a, b)
    elif operator =="subtract":
        return subtract(a, b)
    elif operator =="multiply":
        return multiply(a, b)
    elif operator == "divide":
        return divide(a, b)
    
    
print(calculator(2, 3))

# add(2,3)
```

```python
# functions returning functions

def create_func(op="add"):
    def add(a, b):
        return a + b

    def subtract(a, b):
        return a - b

    def multiply(a, b):
        return a * b

    def divide(a, b):
        return a / b
    
    if op == "add":
        return add
    elif op =="subtract":
        return subtract
    elif op =="multiply":
        return multiply
    elif op == "divide":
        return divide
    
add_fn = create_func()
print(add_fn(10, 2))
print(add_fn(10, 20))

subtract_fn = create_func("subtract")
print(subtract_fn(10, 2))
```

```python
# functions as arguments

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    return a / b

def quick_calc(a, b, op):
    return op(a, b)


quick_calc(10, 2, multiply)
```

#### Intro to Decorators

```python
# function working with strings

def modify(string_in):
    string_out = string_in + "!"
    return string_out

```

```python
# function working with functions instead of strings

def decorator(func):
    def wrapped_func():
        print("Doing some work before executing function...")
        func()
        print("Doing some work after executing function...")
    return wrapped_func
    
# either create new wrapped function


# or overwrite original function as a wrapped function

```

#### Decorators - Examples

```python
def print_log(func):
    def func_with_log(a, b):
        print("-----")
        print("Calling function...")
        result = func(a, b)
        print(f"The result is {result}")
        print("-----")
        return result  # note: return whatever func returns
    return func_with_log

```

```python
import time

def time_it(func):
    def new_func():
        start = time.time()
        result = func()
        end = time.time()
        elapsed = end - start
        print(f"Function took {elapsed} seconds")
        return result  # note: return whatever func returns
    return new_func

# time a list comprehension


# time a for loop

```
